## Pre-processing data

```python
# creating datetime col
buoy['datetime'] = pd.to_datetime(buoy['Datetime'])
buoy.drop(columns='Datetime',inplace=True)
buoy.set_index('datetime',inplace=True)

# creating julian days col
jd_2022 = pd.to_datetime('2022-01-01T00:00').to_julian_date()
jd_data_start = pd.to_datetime(buoy.index[0]).to_julian_date()
buoy['jd_2022'] = buoy.index.to_julian_date() - jd_2022
buoy['jd_dstart'] = buoy.index.to_julian_date() - jd_data_start

# selecting relevant data
buoy = buoy[['swvht1','tp1','wvdir1','jd_2022', 'jd_dstart']]
buoy = buoy.rename(columns={'swvht1':'Hs','tp1':'Tp','wvdir1':'Dm'})
# buoy['z'] = 0
buoy.reset_index(inplace=True) 
```

## Quality Control (IOOS)

The aim of this session is to promote a series of quality control tests on the data following the IOOS Manual for Real-Time Quality Control of Water Level Data.

The applied tests are:
- Gross Range test
- Spike test
- Rate of Change test
- Flat Line test

```python
# Configurate the tests
import ioos_qc
from ioos_qc.config import QcConfig

# Significant Height (Hs)
roc_threshold = buoy['Hs'].std()*3

qc_config_hs = {
       "qartod": {
           
      "gross_range_test": {
        "fail_span": [0, 19.9],
        "suspect_span": [0.1, 7]
      },
      "flat_line_test": {
        "tolerance": 0.001,
        "suspect_threshold": 10800,
        "fail_threshold": 21600
      },
      "spike_test": {
        "suspect_threshold": 0.8,
        "fail_threshold": 3,
      },
        "rate_of_change_test": {
            "threshold":roc_threshold
        }
           
    }
}

# Peak Period (Tp)
roc_threshold = buoy['Tp'].std()*3

qc_config_tp = {
    "qartod": {
      "gross_range_test": {
        "fail_span":  [1.7, 30],
        "suspect_span": [2, 28]
      },
      "flat_line_test": {
        "tolerance": 0.001,
        "suspect_threshold": 10800,
        "fail_threshold": 21600
      },
      "spike_test": {
       "threshold": 3,
      },
        "rate_of_change_test": {
        "threshold": roc_threshold, 
       }
        
    }
}

# Peak Period (Dm)
roc_threshold = buoy['Dm'].std()*3

qc_config_dm = {
    "qartod": {
      "gross_range_test": {
        "fail_span":  [0, 360],
        "suspect_span": [1, 359]
      },
      "flat_line_test": {
        "tolerance": 0.001,
        "suspect_threshold": 10800,
        "fail_threshold": 21600
      },
      "spike_test": {
       "threshold": 3,
      },
        "rate_of_change_test": {
        "threshold": roc_threshold, 
       }
        
    }
}
```

```python
qc_hs = QcConfig(qc_config_hs)
qc_tp = QcConfig(qc_config_tp)
qc_dm = QcConfig(qc_config_dm)
```

```python
def concat_flags_to_data(data_original, qc_param, param=None):
   
    data = data_original.copy()
    
    qc_results =  qc_param.run(
    inp=buoy[param],
    tinp=buoy["datetime"])
    # zinp=buoy["z"])
    
    grt = qc_results['qartod']['gross_range_test']
    spkt = qc_results['qartod']['spike_test']
    flt = qc_results['qartod']['flat_line_test']
    roct = qc_results['qartod']['rate_of_change_test']
    tests = np.stack([grt,spkt,flt,roct]).T
    
    cols_names = [f'{param}_gross_range_test',f'{param}_spike_test',f'{param}_flat_line_test',f'{param}_rate_of_change_test']
    
    flags_df = pd.DataFrame(tests,columns=cols_names,index=data.index)
     
    return flags_df,qc_results
```

```python
qc_results =  qc_hs.run(
    inp=buoy['Hs'],
    tinp=buoy["datetime"])
```

```python
qc_results['qartod']['gross_range_test']
```

```python
flags_hs,qc_results_hs = concat_flags_to_data(buoy, qc_hs, param='Hs')
flags_tp,qc_results_tp = concat_flags_to_data(buoy, qc_tp, param='Tp')
buoy_flags = pd.concat([buoy,flags_hs,flags_tp],axis=1)
```

```python
from datetime import datetime

import numpy as np
import matplotlib.pyplot as plt

def plot_results(data, var_name, results, title, test_name,ax):
    time = data["datetime"]
    obs = data[var_name]
    qc_test = results["qartod"][test_name]

    qc_pass = np.ma.masked_where(qc_test != 1, obs)
    qc_suspect = np.ma.masked_where(qc_test != 3, obs)
    qc_fail = np.ma.masked_where(qc_test != 4, obs)
    qc_notrun = np.ma.masked_where(qc_test != 2, obs)

    # fig, ax = plt.subplots(figsize=(15, 3.75))
    # fig.set_title = f"{test_name}: {title}"
    
    # ax.set_xlabel("Time")
    ax.set_ylabel("Observation Value")
    ax.set_title(f'{test_name}')
    
    kw = {"marker": "o", "linestyle": "none"}
    ax.plot(time, obs,  label="obs", color="#A6CEE3")
    ax.plot(time, qc_notrun, markersize=2, label="qc not run", color="gray", alpha=0.2, **kw)
    ax.plot(time, qc_pass, markersize=4, label="qc pass", color="green", alpha=0.5, **kw)
    ax.plot(time, qc_suspect, markersize=4, label="qc suspect", color="orange", alpha=0.7, **kw)
    ax.plot(time, qc_fail, markersize=6, label="qc fail", color="red", alpha=1.0, **kw)
    ax.grid(True)
    
        
    import matplotlib.ticker as mticker
    import matplotlib.dates as mdates

    ticks_loc = ax.get_xticks().tolist()
    ax.xaxis.set_major_locator(mticker.FixedLocator(ticks_loc))    
    ax.set_xticklabels(ticks_loc, rotation=0, ha='center')
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m'))

```

```python
fig,ax = plt.subplots(4,1,sharex=True,figsize=(10,8))

tests_names = ['gross_range_test','spike_test','flat_line_test','rate_of_change_test']

for i,test_name in enumerate(tests_names):
    plot_results(
        buoy,
        "Hs",
        qc_results_hs, "title",
        test_name,
        ax=ax[i]
    )
ax[i].set_xlabel("Time");
```

```python
fig,ax = plt.subplots(4,1,sharex=True,figsize=(10,13))

tests_names = ['gross_range_test','spike_test','flat_line_test','rate_of_change_test']

for i,test_name in enumerate(tests_names):
    plot_results(
        buoy,
        "Tp",
        qc_results_tp, "title",
        test_name,
        ax=ax[i]
    )
```

```python
fig,ax = plt.subplots(4,1,sharex=True,figsize=(10,13))

tests_names = ['gross_range_test','spike_test','flat_line_test','rate_of_change_test']

for i,test_name in enumerate(tests_names):
    plot_results(
        buoy,
        "Dm",
        qc_results_tp, "title",
        test_name,
        ax=ax[i]
    )
```

## Statistical Indicators

- Root Mean Square Error (RMSE)
- Simetric Slope (SS)
- Mean bias (bias)
- Scatter indes (SI)

```python
params = ['Hs','Tp','Dm']
N = len(buoy)
columns = ['RMSE','Bias','Si','SS','N']
table_stats = pd.DataFrame(columns=columns,index=params).round(2)
```

```python
for param in params:
    rmse = np.sqrt(np.square(waves[param].values - buoy_flags[param]).mean())
    ss = np.sqrt( np.sum(waves[param].values**2)/np.sum(buoy_flags[param]**2))
    bias = waves[param].values.mean() - buoy_flags[param].mean()
    si = rmse / buoy_flags[param].mean()
    values = [rmse,bias,si,ss,N]
    index = param
    table_stats.loc[param] = np.array(values).round(2)
table_stats['N'] = table_stats['N'].astype(int)
    
    
```

```python
table_stats
```

```python
# relevant statistical parameters
ss_hs = table_stats.loc['Hs','SS']
ss_tp = table_stats.loc['Tp','SS']
ss_dm = table_stats.loc['Dm','SS']

fig, ax = plt.subplots(1,3,figsize=(14,4));
plt.subplots_adjust(wspace=0.3)

ax[0].plot(buoy_flags['Hs'],waves['Hs'],'.k')
ax[1].plot(buoy_flags['Tp'],waves['Tp'],'.k')
ax[2].plot(buoy_flags['Dm'],waves['Dm'],'.k')

ax[0].plot([0, 1], [0, 1], transform=ax[0].transAxes,color='grey',lw=1,ls='--',alpha=0.8);
ax[0].plot([0,1], [0,1*ss_hs], transform=ax[0].transAxes,color='red',lw=1,ls='--',alpha=0.8);

ax[1].plot([0, 1], [0, 1], transform=ax[1].transAxes,color='grey',lw=1,ls='--');
ax[1].plot([0,1], [0,1*ss_tp], transform=ax[1].transAxes,color='red',lw=1,ls='--',alpha=0.8);

ax[2].plot([0, 1], [0, 1], transform=ax[2].transAxes,color='grey',lw=1,ls='--');
ax[2].plot([0,1], [0,1*ss_dm], transform=ax[2].transAxes,color='red',lw=1,ls='--',alpha=0.8);

lims = [(0.8,2.6), (4,24), (0,360)]
for i,lim in enumerate(lims):
    ax[i].set_ylim(lim)
    ax[i].set_xlim(lim)
# ax[0].axis('equal')

ax[0].set_xlabel('Buoy Hs (m)')
ax[0].set_ylabel('Model Hs (m)')
ax[1].set_xlabel('Buoy Tp (s)')
ax[1].set_ylabel('Model Tp (s)');
ax[2].set_xlabel('Buoy Dm (°)')
ax[2].set_ylabel('Model Dm (°)');
```
